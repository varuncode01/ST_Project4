```python
from sklearn.datasets import load_breast_cancer
data = load_breast_cancer()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.DataFrame(data.data, columns=data.feature_names)
df['target'] = pd.Series(data.target)
df.head()
```

```python
df.shape
```

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
df.info()
```

```python
df.value_counts().sum()
```

```python
df.columns
```

```python
selected_features = ['mean radius', 'mean texture', 'mean perimeter', 'mean area',
       'mean smoothness', 'mean compactness', 'mean concavity',
       'mean concave points', 'mean symmetry', 'mean fractal dimension',
       'worst radius', 'worst texture', 'worst perimeter', 'worst area',
       'worst smoothness', 'worst compactness', 'worst concavity',
       'worst concave points', 'worst symmetry', 'worst fractal dimension']
```

```python
x = df.drop(columns = ['radius error', 'texture error', 'perimeter error', 'area error',
       'smoothness error', 'compactness error', 'concavity error',
       'concave points error', 'symmetry error', 'fractal dimension error', 'target'])
y = df['target']
```

```python
x.shape
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
x_train.shape
```

```python
model = Sequential()
model.add(Dense(20,activation='relu',input_dim=20))   # Input
model.add(Dense(20,activation='relu'))             # hidden - 1
model.add(Dense(40,activation='relu'))             # hidden - 2
model.add(Dense(20,activation='relu'))             # hidden - 2
model.add(Dense(1,activation='sigmoid'))              # output
```

```python
model.summary()
```

```python
model.compile(optimizer='Adam',loss = 'binary_crossentropy',metrics = ['accuracy'])
```

```python
model.fit(x_train,y_train,epochs=20,validation_split = 0.2)
```

```python
train_pred = model.predict(x_train)
train_pred
```

```python
train_pred = np.where(train_pred>0.5,1,0)
print(train_pred)
```

```python
test_pred = model.predict(x_test)
test_pred = np.where(test_pred>0.5,1,0)
test_pred
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
```

```python
print('Train acc :- ', accuracy_score(y_train,train_pred))
print('Test confusion_matrix :- \n', confusion_matrix(y_test,test_pred))
```

```python
print('Test acc :- ', accuracy_score(y_test,test_pred))
print('Test confusion_matrix :- \n', confusion_matrix(y_test,test_pred))
```

```python
y_pred_prob = model.predict(x_test)  # Gives probabilities
```

```python
from sklearn.metrics import precision_score, recall_score
```

```python
y_pred = (y_pred_prob > 0.5).astype("int32")  # Threshold at 0.5
```

```python
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)

print("Precision:", precision)
print("Recall:", recall)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```
